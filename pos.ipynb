```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
url = 'https://graphql.stakingrewards.com/?operationName=GetTableAssets&variables=%7B%22items%22%3A20%2C%22date%22%3A%222022-09-22%22%2C%22skip%22%3A0%2C%22isLaunched%22%3Atrue%2C%22sort%22%3A%22totalValueLocked_DESC%22%2C%22filter%22%3A%22proof-of-stake%22%7D&extensions=%7B%22persistedQuery%22%3A%7B%22version%22%3A1%2C%22sha256Hash%22%3A%229d3b9976b5d5c738b985980558f68c3d6842a58ed9b7abe0c8e41da639b63ced%22%7D%7D'
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header).json()
r = r['data']['assets']

df = pd.DataFrame(r)
df
```

```python
stake_stats = df[['name', 'symbol', 'marketcapUsd', 'totalValueLocked', 'tableTotalStaked']].copy()
stake_stats
```

```python
mean_value_staked_in_USD_no_ETH = stake_stats['totalValueLocked'].drop([0], axis=0).mean()
mean_value_staked_in_USD_no_ETH
```

```python
ETH_value_staked_in_USD = stake_stats.iloc[0, 3]
ETH_to_average_ratio = ETH_value_staked_in_USD/mean_value_staked_in_USD_no_ETH
ETH_to_average_ratio -1
```

```python
SOL_value_staked_in_USD = stake_stats.iloc[1, 3]
ETH_to_SOL_ratio = ETH_value_staked_in_USD/SOL_value_staked_in_USD
ETH_to_SOL_ratio -1
```

```python
mean_supply_pct_staked_no_ETH = stake_stats['tableTotalStaked'].drop([0], axis=0).mean()
mean_supply_pct_staked_no_ETH
```

```python
with open('params.json') as file:
    parsed_json = json.loads(file.read())

r2 = requests.get('https://api.glassnode.com/v1/metrics/supply/current', params=parsed_json['posParams'])

eth_supply = pd.read_json(r2.text, convert_dates=['t'])
eth_supply
```

```python
eth_actual_supply = eth_supply.iloc[len(eth_supply['v'])-1, 1]
```

```python
projected_staked_supply = (mean_supply_pct_staked_no_ETH/100)*eth_actual_supply
projected_staked_supply
```

```python
projected_number_of_validators = round(projected_staked_supply/32)
projected_number_of_validators
```

```python
projected_issuance_yield = 32/np.sqrt(projected_number_of_validators)
print(f'{round(projected_issuance_yield*100, 2)}%')
```

```python
fee_rev = pd.read_json('fee_rev.json')
fee_rev['projected_fee_yield'] = fee_rev['annualizedFeeRev']/projected_staked_supply
fee_rev['projected_total_yield'] = fee_rev['projected_fee_yield'] + projected_issuance_yield
fee_rev
```
