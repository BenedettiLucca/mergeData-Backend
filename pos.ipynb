```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://graphql.stakingrewards.com/?operationName=GetTableAssets&variables={%22items%22:20,%22date%22:%222022-09-21%22,%22skip%22:0,%22isLaunched%22:true,%22sort%22:%22totalValueLocked_DESC%22,%22filter%22:%22proof-of-stake%22}&extensions={%22persistedQuery%22:{%22version%22:1,%22sha256Hash%22:%229d3b9976b5d5c738b985980558f68c3d6842a58ed9b7abe0c8e41da639b63ced%22}}'
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header).json()
r = r['data']['assets']

df = pd.DataFrame(r)
df
```

```python
stake_stats = df[['name', 'symbol', 'marketcapUsd', 'totalValueLocked', 'tableTotalStaked']].copy()
stake_stats
```

```python
mean_value_staked_in_USD_no_ETH = stake_stats['totalValueLocked'].drop([0], axis=0).mean()
mean_value_staked_in_USD_no_ETH
```

```python
ETH_value_staked_in_USD = stake_stats.iloc[0, 3]
ETH_to_average_ratio = ETH_value_staked_in_USD/mean_value_staked_in_USD_no_ETH
ETH_to_average_ratio -1
```

```python
SOL_value_staked_in_USD = stake_stats.iloc[1, 3]
ETH_to_SOL_ratio = ETH_value_staked_in_USD/SOL_value_staked_in_USD
ETH_to_SOL_ratio -1
```

```python
mean_supply_pct_staked_no_ETH = stake_stats['tableTotalStaked'].drop([0], axis=0).mean()
mean_supply_pct_staked_no_ETH
```

```python
api_key = '28RLin1KM3Um0BHIjDt6FPDZ1ed'

r2 = requests.get('https://api.glassnode.com/v1/metrics/supply/current', params={'a': 'ETH', 'i': '24h', 'c': 'NATIVE', 'api_key': api_key})

eth_supply = pd.read_json(r2.text, convert_dates=['t'])
eth_supply
```

```python
projected_staked_supply = (mean_supply_pct_staked_no_ETH/100)*eth_actual_supply
projected_staked_supply
```
